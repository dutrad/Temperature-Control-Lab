```python
import numpy as np
from scipy.integrate import odeint
import matplotlib.pyplot as plt
```

```python
#return dy/dt
def model(y,t,k):
    dydt = -k * y
    return dydt

#initial condition
y0 = 5;

#time
t = np.linspace(0,20,50)

#solve
y1 = odeint(model, y0, t, args=(0.1,))
y2 = odeint(model, y0, t, args=(0.2,))
y3 = odeint(model, y0, t, args=(0.5,))

#plot results
plt.plot(t,y1,'r-', linewidth=2, label='k=0.1')
plt.plot(t,y2,'b--', linewidth=2, label='k=0.2')
plt.plot(t,y3,'g:', linewidth=2, label='k=0.5')
plt.grid()
plt.xlabel('time')
plt.ylabel('y(t)')
plt.legend()
plt.title('Exercício 0')
plt.show()
```

```python
#return dy/dt
def model(y,t,a,b):
    dydt = a*y + b
    return dydt

#initial condition
y0 = 0;

#time
t = np.linspace(0,5)

#solve
y = odeint(model, y0, t, args=(-1,1,))

#plot results
plt.plot(t,y,'r-', linewidth=2)
plt.xlabel('time')
plt.ylabel('y(t)')
plt.title('Exercício 1')
plt.grid()
plt.show()
```

```python
#return dy/dt
def model(y,t):
    if(t<10):
        u = 0.0
    else:
        u = 2.0

    dydt = (-y + u)/5
    return dydt

#initial condition
y0 = 0;

#time
t = np.linspace(0,40)

#solve
y = odeint(model, y0, t)

#plot results
plt.plot(t,y,'r-', linewidth=2)
plt.xlabel('time')
plt.ylabel('y(t)')
plt.title('Exercício 2')
plt.grid()
plt.show()
```
